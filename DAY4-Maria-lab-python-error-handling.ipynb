````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory

def get_customer_orders(inventory):
    while True:
        try:
            how_many = int(input("how many orders would you like to place? "))
            if how_many <= 0:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a positive integer for the number of orders.")

    customer_orders = []
    for item in range(how_many):
        while True:
            product_name = input(f"what would you like to order from this list? {products}")
            if product_name in inventory and inventory[product_name] > 0:
                customer_orders.append(product_name)
                break
            else:
                print("Invalid product name or out of stock. Please choose a product from the list.")

    print("You have ordered", how_many, "items: ", customer_orders)
    return customer_orders

def update_inventory(customer_orders, inventory):
    for order in set(customer_orders):
        if order in inventory:
            inventory[order] -= customer_orders.count(order)
    updated_inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return updated_inventory

def calculate_price(customer_orders):
    price = []
    for item in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                value = int(input(f"how much does {item} cost?"))
                if value > 0:
                    price.append(value)
                    valid_input = True
                else:
                    print("Price cannot be negative or 0. Please enter a valid price.")
            except ValueError:
                print("Invalid input. Please enter a valid price.")
    total_price = sum(price)
    return print("total price is: $" + str(total_price))

def calculate_order_statistics(customer_orders, products):
    unique_inventory = len(products)
    unique_ordered = len(set(customer_orders))
    percentage_ordered = round(unique_ordered / unique_inventory * 100, 2)
    order_statistics = (unique_ordered, percentage_ordered)
    return order_statistics

def print_order_statistics(order_statistics):
    return print(order_statistics)

def print_updated_inventory(inventory):
    updated_inventory = inventory
    return print(updated_inventory)

inventory = initialize_inventory(products)

customer_orders = get_customer_orders(inventory)

inventory = update_inventory(customer_orders, inventory)

calculate_price(customer_orders)

order_statistics = calculate_order_statistics(customer_orders, products)

print_order_statistics(order_statistics)

print_updated_inventory(inventory)
````
